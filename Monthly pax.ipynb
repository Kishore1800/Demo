```python
def transform_df(df,df2,df_klia,df_klia2):
    df2.columns = df2.columns.str.strip()
    df.columns = df.columns.str.strip()
#     col_ck = ['','2023', '2022', '%', '', '','Jan-Mac', 'Apr-Jun', 'Jul-Sep', 'Oct-Dec']
#     columns_to_transpose = df.columns.difference(col_ck) 
#     columns_to_transpose_df2 = df2.columns.difference(col_ck)
    df  = df[['International', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul','Aug', 'Sep', 'Oct', 'Nov', 'Dec']]
    df2 = df2[['Domestic', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug','Sep', 'Oct', 'Nov', 'Dec']]
    melted_df = pd.melt(df, id_vars=['International'], var_name='Month', value_name='Value')
    melted_df['Value'] = pd.to_numeric(melted_df['Value'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    melted_df2 = pd.melt(df2, id_vars=['Domestic'], var_name='Month', value_name='Value')
    melted_df2['Value'] = pd.to_numeric(melted_df2['Value'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    melted_df.rename(columns = {'International':'Airport','Value':'International'},inplace = True)
    melted_df2.rename(columns = {'Domestic':'Airport','Value':'Domestic'},inplace = True)
    melted_df['Airport'] = melted_df['Airport'].str.strip()
    melted_df2['Airport'] = melted_df2['Airport'].str.strip()
    final_df = pd.merge(melted_df,melted_df2, on = ['Airport','Month'] , how = 'outer')
    
    df_klia = df_klia.T
    df_klia.columns = df_klia.iloc[1]
    df_klia = df_klia[2:].reset_index()
    df_klia.columns = ['Month', 'International', 'Domestic']
    df_klia['Airport'] = 'KLIA Terminal 1'
    df_klia = df_klia[0:12]
    df_klia2 = df_klia2.T
    df_klia2.columns = df_klia2.iloc[1]
    df_klia2 = df_klia2[2:].reset_index()
    df_klia2.columns = ['Month', 'International', 'Domestic']
    df_klia2['Airport'] = 'KLIA Terminal 2'
    df_klia2 = df_klia2[0:12]
    df_klia['International']= pd.to_numeric(df_klia['International'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    df_klia['Domestic']= pd.to_numeric(df_klia['Domestic'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    df_klia2['International']= pd.to_numeric(df_klia2['International'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    df_klia2['Domestic']= pd.to_numeric(df_klia2['Domestic'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    final_df = pd.merge(final_df[final_df['Airport']!='KLIA'],pd.merge(df_klia,df_klia2,on=['Airport','Month','International','Domestic'],how = 'outer'),on=['Airport','Month','International','Domestic'],how = 'outer')
#     final_df = final_df.sort_values(by = 'Airport')
    final_df.reset_index(drop = True,inplace = True)
#     final_df['International']= pd.to_numeric(final_df['International'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
#     final_df['Domestic']= pd.to_numeric(final_df['Domestic'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    final_df['Subs_category'] = final_df['Airport'].apply(lambda x: 'KUL' if x in ['KLIA Terminal 1', 'KLIA Terminal 2'] else 'MASB')
    final_df['Month'] = final_df['Month'].str.strip()
    return final_df
```

```python
def transform_budget(df_budget,df_budget2):
    df_budget = df_budget[['International', 'January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']]
    df_budget.columns = [df_budget.columns[0]] + [col[:3] for col in df_budget.columns[1:]]
    df_budget['International'] = df_budget['International'].str.strip()
    df_budget = df_budget[~df_budget['International'].isin(['ASEAN','Non-ASEAN',''])]
    df_budget2 = pd.DataFrame(data_budget[149:204], columns=data_budget[146])
    df_budget2.columns = df_budget2.columns.str.strip()
    df_budget2 = df_budget2[['Domestic', 'January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']]
    df_budget2.columns = [df_budget2.columns[0]] + [col[:3] for col in df_budget2.columns[1:]]
    df_budget2['Domestic'] = df_budget2['Domestic'].str.strip()
    df_budget2 = df_budget2[~df_budget2['Domestic'].isin(['ASEAN','Non-ASEAN',''])]
    melted_df = pd.melt(df_budget, id_vars=['International'], var_name='Month', value_name='Value')
    melted_df2 = pd.melt(df_budget2, id_vars=['Domestic'], var_name='Month', value_name='Value')
    melted_df.rename(columns = {'International':'Airport','Value':'International_Budget'},inplace = True)
    melted_df2.rename(columns = {'Domestic':'Airport','Value':'Domestic_Budget'},inplace = True)
    final_df = pd.merge(melted_df,melted_df2, on = ['Airport','Month'] , how = 'right')
    final_df['Airport'] = final_df['Airport'].str.strip()
    final_df['Month'] = final_df['Month'].apply(lambda x: datetime.strptime(x, '%b').strftime('2024-%m-01'))
    final_df.rename(columns = {'Month':'Date'},inplace = True)
    final_df['Date'] = pd.to_datetime(final_df['Date'])
    final_df['International_Budget']= pd.to_numeric(final_df['International_Budget'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    final_df['Domestic_Budget']= pd.to_numeric(final_df['Domestic_Budget'].str.replace(',', '').str.replace(' ', ''), errors='coerce').fillna(0)
    final_df['Airport'].replace({'Terminal 1':'KLIA Terminal 1','Terminal 2':'KLIA Terminal 2'},inplace = True)
    return final_df
```

```python
df_final_budget = transform_budget(df_budget,df_budget2)
df_final_budget
```
